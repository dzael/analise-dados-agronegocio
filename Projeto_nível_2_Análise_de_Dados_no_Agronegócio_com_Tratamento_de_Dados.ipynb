```python
# Análise de Dados no Agronegócio com Tratamento de Dados: Projeto focado na limpeza, organização e análise de dados agrícolas, simulando cenários reais com dados incompletos.

# Objetivo: Analisar dados de produção agrícola contendo inconsistências, realizando o tratamento das informações para garantir análises confiáveis.

import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados

dados = {
    "Ano": [2020, 2021, 2022, 2020, 2021, None],
    "Cultura": ["Soja", "Soja", "Soja", "Milho", None, "Milho"],
    "Área (hectares)": [100, 110, None, 80, 90, 100],
    "Produção (toneladas)": [320, None, 400, 200, 260, 300]
}

df = pd.DataFrame(dados)
df

# Tratamento de dados

df = df.dropna()
df

# Análise dos dados após tratamento

producao_por_ano = df.groupby('Ano')['Produção (toneladas)'].sum()

producao_por_ano.plot(kind='bar')
plt.title('Produção por Ano (dados tratados)')
plt.xlabel('Ano')
plt.ylabel('Produção')
plt.show()

producao_por_cultura = df.groupby('Cultura')['Produção (toneladas)'].sum()

producao_por_cultura.plot(kind='bar')
plt.title('Produção por Cultura (dados tratados)')
plt.xlabel('Cultura')
plt.ylabel('Produção')
plt.show()

# Análise de produtividade

df['Produtividade (t/ha)'] = df['Produção (toneladas)'] / df['Área (hectares)']
df

df.groupby('Cultura')['Produtividade (t/ha)'].mean().plot(kind='bar')
plt.title('Produtividade média por cultura')
plt.xlabel('Cultura')
plt.ylabel('t/ha')
plt.show()
```
